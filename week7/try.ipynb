```python
import unicodedata
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from typing import List

# One-time setup (downloads if missing)
def _ensure_nltk():
    try:
        nltk.data.find("tokenizers/punkt")
    except LookupError:
        nltk.download("punkt", quiet=True)
    try:
        nltk.data.find("corpora/stopwords")
    except LookupError:
        nltk.download("stopwords", quiet=True)

def tokenize_text(
    text: str,
    *,
    lowercase: bool = True,
    strip_accents: bool = False,
    remove_stopwords: bool = False,
    language: str = "english",
) -> List[str]:
    _ensure_nltk()

    # --- Normalize Unicode ---
    s = unicodedata.normalize("NFKC", text)
    s = (s.replace("—", "-").replace("–", "-")
           .replace("“", '"').replace("”", '"')
           .replace("‘", "'").replace("’", "'"))

    if lowercase:
        s = s.casefold()

    if strip_accents:
        s = "".join(ch for ch in unicodedata.normalize("NFD", s)
                    if unicodedata.category(ch) != "Mn")

    # --- Tokenize ---
    tokens = word_tokenize(s, language=language)

    # --- Filter punctuation-only tokens ---
    tokens = [t for t in tokens if any(ch.isalnum() for ch in t)]

    # --- Stopword removal (optional) ---
    if remove_stopwords:
        stop_words = set(stopwords.words(language))
        tokens = [t for t in tokens if t not in stop_words]

    return tokens
```

```python
from dataclasses import dataclass, field
from typing import Dict, List

@dataclass
class Posting:
    """Per-document stats for a term."""
    tf: int = 0
    positions: List[int] = field(default_factory=list)

@dataclass
class TermEntry:
    """One term's global entry in the inverted index."""
    document_frequency: int = 0        # df: number of docs containing the term
    total_frequency: int = 0           # cf: total occurrences across all docs
    documents: Dict[str, Posting] = field(default_factory=dict)  # docID -> Posting

    def add_positions(self, doc_id: str, positions: List[int]) -> None:
        """Add one or many positions for this term in a given document."""
        if not positions:
            return
        n = len(positions)
        post = self.documents.get(doc_id)
        if post is None:
            # first time this doc contains the term
            self.documents[doc_id] = Posting(tf=n, positions=list(positions))
            self.document_frequency += 1
            self.total_frequency += n
        else:
            post.tf += n
            post.positions.extend(positions)
            self.total_frequency += n

@dataclass
class Vocabulary:
    """The inverted index: term -> TermEntry."""
    vocab: Dict[str, TermEntry] = field(default_factory=dict)

    def add_term(self, term: str, doc_id: str, positions: List[int]) -> None:
        """
        Merge all occurrences of `term` from one document.
        `positions` must be 0-based, ascending positions in the doc's token stream.
        """
        entry = self.vocab.get(term)
        if entry is None:
            entry = TermEntry()
            self.vocab[term] = entry
        entry.add_positions(doc_id, positions)
```

```python
from typing import Callable, Dict, List, Optional
from collections import defaultdict

# Expecting you already have:
# - tokenize_text(text: str) -> List[str]     # your NLTK-based analyzer
# - class Vocabulary with .add_term(term, doc_id, positions)

def index_text_into_vocab(
    vocab: "Vocabulary",
    doc_id: str,
    text: str,
    *,
    analyze: Callable[[str], List[str]] = None,
    doc_store: Optional[Dict[str, Dict[str, int]]] = None,
) -> None:
    tokens = analyze(text)  # IMPORTANT: use the SAME analyzer for indexing & querying

    if not tokens:
        if doc_store is not None:
            doc_store[doc_id] = {"length": 0}
        return

    # Build per-doc map: term -> [positions...]
    term_positions: Dict[str, List[int]] = defaultdict(list)
    for i, tok in enumerate(tokens):
        term_positions[tok].append(i)

    # Update global vocab
    for term, pos_list in term_positions.items():
        vocab.add_term(term, doc_id, pos_list)

    # Optional: record doc length for scoring (e.g., BM25)
    if doc_store is not None:
        doc_store[doc_id] = {"length": len(tokens)}

```

```python
import os, json, hashlib
from typing import Dict, TextIO, List

def stable_partition(term: str, R: int) -> int:
    return int(hashlib.md5(term.encode("utf-8")).hexdigest(), 16) % R

def shuffle_path(mapper_id: int, rid: int, root: str) -> str:
    # by-reducer layout
    return os.path.join(root, f"part-{rid:03d}", f"map-{mapper_id:05d}.jsonl")

class ShuffleWriter:
    def __init__(self, mapper_id: int, R: int, root: str):
        self.mapper_id, self.R, self.root = mapper_id, R, root
        self._handles: Dict[int, TextIO] = {}

    def _handle(self, rid: int) -> TextIO:
        path = shuffle_path(self.mapper_id, rid, self.root)
        os.makedirs(os.path.dirname(path), exist_ok=True)
        if rid not in self._handles:
            self._handles[rid] = open(path, "a", encoding="utf-8")
        return self._handles[rid]

    def write(self, term: str, doc_id: str, positions: List[int]) -> None:
        rid = stable_partition(term, self.R)
        fh = self._handle(rid)
        fh.write(json.dumps({"t": term, "d": doc_id, "p": positions, "tf": len(positions)}, ensure_ascii=False) + "\n")

    def close(self) -> None:
        for fh in self._handles.values():
            fh.flush(); fh.close()
        self._handles.clear()

def emit_mapper_vocab(mapper_id: int, vocab: "Vocabulary", *, R: int, shuffle_root: str) -> None:
    sw = ShuffleWriter(mapper_id, R, shuffle_root)
    for term, entry in vocab.vocab.items():
        for doc_id, posting in entry.documents.items():
            sw.write(term, doc_id, posting.positions)
    sw.close()
```

```python
import os, json
from collections import defaultdict
from typing import Dict, List

def reduce_partition(rid: int, shuffle_root: str, segments_root: str) -> str:
    part_dir = os.path.join(shuffle_root, f"part-{rid:03d}")
    os.makedirs(segments_root, exist_ok=True)

    term_docs: Dict[str, Dict[str, Dict[str, List[int] | int]]] = defaultdict(dict)
    cf_accum: Dict[str, int] = defaultdict(int)

    if os.path.isdir(part_dir):
        for name in sorted(os.listdir(part_dir)):
            path = os.path.join(part_dir, name)
            with open(path, "r", encoding="utf-8") as fh:
                for line in fh:
                    s = line.strip()
                    if not s: continue
                    rec = json.loads(s)
                    t, d, p = rec["t"], rec["d"], rec["p"]
                    tf = int(rec.get("tf", len(p)))
                    bucket = term_docs[t].get(d)
                    if bucket is None:
                        term_docs[t][d] = {"frequency": tf, "positions": list(p)}
                    else:
                        bucket["frequency"] += tf
                        bucket["positions"].extend(p)
                    cf_accum[t] += tf

    # finalize & write
    terms_out = {}
    for t, docs in term_docs.items():
        for d, rec in docs.items():
            rec["positions"] = sorted(int(x) for x in rec["positions"])
            rec["frequency"] = int(rec["frequency"])
        terms_out[t] = {
            "document_frequency": len(docs),
            "total_frequency": int(cf_accum[t]),
            "documents": {d: docs[d] for d in sorted(docs.keys())},
        }

    out_path = os.path.join(segments_root, f"reduce-{rid:03d}-segment.json")
    with open(out_path, "w", encoding="utf-8") as f:
        json.dump({"terms": terms_out}, f, ensure_ascii=False, indent=2)
    return out_path
```

```python
def vocab_to_dict(vocab: "Vocabulary", limit_docs: Optional[int] = None):
    out = {}
    for term, entry in vocab.vocab.items():
        docs = {}
        for i, (doc_id, p) in enumerate(sorted(entry.documents.items())):
            if limit_docs is not None and i >= limit_docs:
                break
            docs[doc_id] = {"frequency": p.tf, "positions": list(p.positions)}
        out[term] = {
            "document_frequency": entry.document_frequency,
            "total_frequency": entry.total_frequency,
            "documents": docs,
        }
    return out
```

```python
from typing import List

def reduce_all(R: int, shuffle_root: str, segments_root: str) -> List[str]:
    return [reduce_partition(rid, shuffle_root, segments_root) for rid in range(R)]
```

```python
import os, json, shutil
from math import ceil
from multiprocessing import Pool, cpu_count
from invertedIndex import mapper_emit_worker, reduce_partition, stable_partition

# ---- config ----
R = 4
M = min(cpu_count(), 4)           # number of mapper processes you want
shuffle_root = "shuffle"
segments_root = "segments"

# (optional) start clean to avoid appending to old shuffle files:
for d in (shuffle_root, segments_root):
    shutil.rmtree(d, ignore_errors=True)

# Build (doc_id, content) rows and split into M batches
rows = list(docs[['id','content']].itertuples(index=False, name=None))
batch_size = ceil(len(rows) / max(M, 1)) or 1
batches = [rows[i:i+batch_size] for i in range(0, len(rows), batch_size)]

# ---- MAP + SHUFFLE (parallel) ----
with Pool(processes=len(batches)) as pool:
    done = pool.starmap(
        mapper_emit_worker,
        [(mid, batch, R, shuffle_root) for mid, batch in enumerate(batches)]
    )
print("Mappers finished:", done)

# ---- REDUCE ALL PARTITIONS (parallel or serial) ----
# Parallel:
with Pool(processes=min(R, cpu_count())) as pool:
    segment_paths = pool.starmap(
        reduce_partition,
        [(rid, shuffle_root, segments_root) for rid in range(R)]
    )
print("Segments written:", segment_paths)

# Optional: sanity check a term, e.g., "genomic"
term = "genomic"
rid = stable_partition(term, R)
seg_path = os.path.join(segments_root, f"reduce-{rid:03d}-segment.json")
with open(seg_path, "r", encoding="utf-8") as f:
    seg = json.load(f)
print(f'Posting for "{term}":\n', json.dumps(seg["terms"].get(term, {}), indent=2))
```
